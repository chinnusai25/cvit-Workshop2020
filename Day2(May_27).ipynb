**Ex 5:**<br>
Write a function which takes an image section and filter as an input and gives <br>the convolution output. Take the filter to be- np.array([[1,1,1],[1,1,1],[1,1,1]])/9


```
# Write a function which takes an image section and filter as an input and gives
# the convolution output. Take the filter to be- np.array([[1,1,1],[1,1,1],[1,1,1]])/9
def convulated_output(H,I):
  # I = cv2.imread('/content/sudoku.jpeg',cv2.IMREAD_GRAYSCALE)
  I1 = I
  for i in range(1,len(I)-1):
    for j in range(1,len(I[0])-1):
      temp=0
      for m in range(-1,2):
        for n in range(-1,2):
          temp+=I[i+m][j+n]*H[1+m,1+n]
      I[i][j]=temp
  cv2_imshow(np.hstack([I1,I]))

I = cv2.imread('/content/sudoku.jpeg',cv2.IMREAD_GRAYSCALE)
H = np.ones((3,3))/9
# print(I.shape)
convulated_output(H,I)
```

**Ex 6:** <br>
Count the number of windows in image 'window.jpeg' using edge detection filters.


```
# Count the number of windows in image 'window.jpeg' using edge detection filters.

I = cv2.imread('/content/window.jpeg',cv2.IMREAD_GRAYSCALE)
H1 = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
H2 = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
I1=np.zeros((len(I),len(I[0])))
I2=np.zeros((len(I),len(I[0])))

for i in range(1,len(I)-1):
  for j in range(1,len(I[0])-1):
    temp=0
    for m in range(-1,2):
      for n in range(-1,2):
        temp+=I[i+m][j+n]*H1[1+m,1+n]
    I1[i][j]=temp

for i in range(1,len(I)-1):
  for j in range(1,len(I[0])-1):
    temp=0
    for m in range(-1,2):
      for n in range(-1,2):
        temp+=I[i+m][j+n]*H2[1+m,1+n]
    I2[i][j]=temp

I = np.sqrt(np.square(I1.astype('int')) + np.square(I2.astype('int')))
cv2_imshow(np.hstack([I1,I2,I]))

count1=0
count2=0

for i in range(len(I1)): 
  temp=0
  for j in range(len(I1[0])):
    if(I1[i][j]>=200 and I1[i][j-1]<200):
      temp+=1
      if(count1<=temp):
        count1=temp

for i in range(len(I2.T)): 
  temp=0
  for j in range(len(I2.T[0])):
    if(I2.T[i][j]>=200 and I2.T[i][j-1]<200):
      temp+=1
      if(count2<=temp):
        count2=temp

print(count1,count2)
print(count1*count2)
```
