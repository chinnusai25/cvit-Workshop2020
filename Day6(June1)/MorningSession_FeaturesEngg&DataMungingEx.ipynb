### TASK 1: Read the different fields from the raw data and create a list of list with appropriate datatypes for each column.

```
def read_data(filename):
  final_data = []
  firstLine = True
  data = []
  for line in open(filename):
    if firstLine:
        firstLine = False
    else:
        data.append(line)
  # Add code here
  for i in range(len(data)):
    temp=data[i].split(",")
    temp[0]=temp[0].replace('"', '')
    temp[1]=float(temp[1])
    temp[2]=float(temp[2])
    final_data.append([temp[0],temp[1],temp[2]])
  return final_data

data = read_data(FULL_DATA)
print(data)
# print(type(data[0][0]),type(data[0][1]),type(data[0][2]))
# NOTE: Data should be of format: [["Male", 73.23, 241.34], ["Female", 56.32, 154.344], ....]
```

Now we need to convert inches to cm, pounds to kg and round these to the nearest integer and we are done. Here is the final code to do the same.

```
def read_data_2(filename):
  final_data = []
  # Add code here, but also convert inches to cm and pounds to kg and round to nearest integer while reading.
  #NOTE: 1 inch = 2.54 cm, 1 pound = 0.4536 kg
  final_data = []
  firstLine = True
  data = []
  for line in open(filename):
    if firstLine:
        firstLine = False
    else:
        data.append(line)
  # Add code here
  for i in range(len(data)):
    temp=data[i].split(",")
    temp[0]=temp[0].replace('"', '')
    temp[1]=round(float(temp[1])*2.54)
    # temp[1]=round(float(temp[1])*2.54,3) #rounding to 3 decimal
    temp[2]=round(float(temp[2])*0.4536)
    # temp[2]=round(float(temp[2])*0.4536,3) #rounding to 3 decimal
    final_data.append([temp[0],temp[1],temp[2]])
  return final_data

data = read_data_2(FULL_DATA)
print(data)
```
