Task 2.2: Remove key 100 from the dict <br> 
Task 2.3: Convert the above dict into a list of key, value tuple. Then sort the list based on mod of the key value with number 101


```
x = dict([(pow(10,r),r) for r in range(10)])
print("(Task 2.1)Keys in the Dict: ",x.keys())
del x[100]
print("(Task 2.2)After deleting key 100: ",x.keys())
listx = [(k, v) for k, v in x.items()] 
print("Dict to list:",listx)
def takeKey(elem):
    return elem[0]%101
listx.sort(key=takeKey)
print("(Task 2.3)After sorting wrt key%101: ",listx)
```

Can you write the relation between tanh and sigmoid functions?

Relation between Tanh and Sigmoid function: <br>
The hyperbolic tangent (tanh) and sigmoid ($\sigma$) functions are defined as follows:

**sigmoid function**
\begin{equation}
sigma(x) = \frac{1}{1 + e^{-x}}
\end{equation}

**tanh function**
\begin{equation}
tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
\end{equation}

Let us plot the both in a single graph:

```
X = np.linspace(-4,4,100)
y_sigma = (1/(1+np.exp(-X)))
y_tanh = (np.exp(X)-np.exp(-X))/(np.exp(X)+np.exp(-X))
plt.plot(X,y_sigma)
plt.plot(X,y_tanh)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend(["Sigmanoid","Tanh"])
```

Let us find the mathematical relation between them,
We know that, 
\begin{equation}
{1}-\frac{1}{1 + e^{-x}} = \frac{1}{1 + e^{x}}
\end{equation}
therefore 
\begin{equation}
1 - \sigma(x) = \sigma(-x)
\end{equation}
Now, 
\begin{equation}
tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}}
\end{equation}
<br>
\begin{equation}
tanh(x) = \frac{e^x + e^{-x} - 2*e^{-x}}{e^x + e^{-x}}
\end{equation}
<br>
\begin{equation}
tanh(x) = 1 - \frac{2*e^{-x}}{e^x + e^{-x}}
\end{equation}
<br>
\begin{equation}
tanh(x) = 1 - \frac{2}{e^{2x} + 1}
\end{equation}
<br>
\begin{equation}
tanh(x) = 1 - 2\sigma(-2x)
\end{equation}
<br>
\begin{equation}
tanh(x) = 1 - 2(1-\sigma(2x))
\end{equation}
<br>
\begin{equation}
tanh(x) = 1 - 2+2\sigma(2x)
\end{equation}
<br>
\begin{equation}
tanh(x) = 2\sigma(2x)-1
\end{equation}
So basically tanh is an example of sigmoid function.

```
r= 10
X = np.linspace(-r,r,100)
y_plus = np.sqrt(r*r-X*X)
y_minus = -np.sqrt(r*r-X*X)
plt.plot(X,y_plus)
plt.plot(X,y_minus)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Circle")
```

```
a= 2
b= 1
Y = np.linspace(-b,b,200)
x_plus = np.sqrt((1-(Y*Y/b*b)))*a
x_minus = -np.sqrt((1-(Y*Y/b*b)))*a
plt.plot(x_plus,Y)
plt.plot(x_minus,Y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Ellipse")
```
